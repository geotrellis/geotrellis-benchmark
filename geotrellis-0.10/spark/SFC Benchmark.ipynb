# Index Benchmarking

We're going to try to answer some questions regarding space filling curve choices as they are implemented in GeoTrellis.
The first question we might ask is of performance. The main job of a space filling curve is to take an arbitrary bounding box and decompose it into a sequence of continous ranges on a line. Since bounding box is a user input we're going to generate a random sample of 1024 bboxes to benchmark zorder, hilbert, and rowmajor space filling curves.

```
for {
  i <- 0 to 1024
} yield {
  val col = scala.util.Random.nextInt(65536)
  val row = scala.util.Random.nextInt(65536)
  val width = scala.util.Random.nextInt(65536)
  val height = scala.util.Random.nextInt(65536)
  KeyBounds(
    SpatialKey(col, row),
    SpatialKey(col+width, row+height))   
}
```

For context a bounding box of 65536 x 65536 tiles sufficient to cover the world somewhere between TMS zoom 12 and 13, resolute enough to see minor streets rendered.

These benchmarks are done on `2.8 GHz Intel Core 2 Duo` processor where all indexing code is single threaded.

````python
## Performance in Time
Lets look at the time to decompose the ranges using each of the curves
````

````python
import pandas as pd
dfz = pd.read_csv("/Users/eugene/z-bench.csv").apply(pd.to_numeric, errors='coerce')
dfh = pd.read_csv("/Users/eugene/hilbert-bench.csv").apply(pd.to_numeric, errors='coerce')
dfr = pd.read_csv("/Users/eugene/rowmajor-bench.csv").apply(pd.to_numeric, errors='coerce')
dfz['area'] = dfz['width']*dfz['height']
dfh['area'] = dfh['width']*dfh['height']
dfr['area'] = dfr['width']*dfr['height']
````

````python
time = dfz.loc[:,('width','height','area')]
time['zorder'] = dfz.loc[:,('nanos')]
time['hilbert'] = dfh.loc[:,('nanos')]
time['rowmajor'] = dfr.loc[:,('nanos')]

ax = time.plot(kind='scatter', x='area', y='hilbert', color='DarkBlue', label='Hilbert', 
    figsize=[12,8], ylim=[0,4e8], xlim=[0,4e9], title="Range Decomposition Time")
time.plot(kind='scatter', x='area', y='zorder', color='DarkGreen', label='Z-Order', ax=ax)
time.plot(kind='scatter', x='area', y='rowmajor', color='DarkRed', label='RowMajor', ax=ax)
ax.set_ylabel("nanos")
````

We can make some general statements. Hilbert is about six times slower to decompose ranges than z-order. However at worst time to full range listing is only 400ms.

````python
time[['area', 'hilbert','zorder','rowmajor']].describe()
````

## Range Count Variability
The count of ranges in decomposition depends on the location of the bounding box as well as its size. An intuition we can use is that the more quadrants a bounding box intercets the more index ranges will be required to cover it. Of course we can not choose where the query bounding box falls, so it is useful to see the effect of its posistion by fixing the area to 1000 x 1000 cells. If each cell represents a tile in a tile layout this is a medium to large query area.

````python
import pandas as pd
dfzf = pd.read_csv("/Users/eugene/z-fixed-area-bench.csv").apply(pd.to_numeric, errors='coerce')
dfhf = pd.read_csv("/Users/eugene/hilbert-fixed-area-bench.csv").apply(pd.to_numeric, errors='coerce')
dfrf = pd.read_csv("/Users/eugene/rowmajor-fixed-area-bench.csv").apply(pd.to_numeric, errors='coerce')

ranges = dfzf.loc[:,('col','row')]
ranges['zorder'] = dfzf.loc[:,('ranges')]
ranges['hilbert'] = dfhf.loc[:,('ranges')]
ranges['rowmajor'] = dfrf.loc[:,('ranges')]

ranges[['hilbert', 'zorder', 'rowmajor']].plot.box(figsize=[10,5]).set_ylabel("ranges")
````
